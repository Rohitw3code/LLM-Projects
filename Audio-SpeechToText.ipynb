```python
import torch

speech_recognizer = pipeline('automatic-speech-recognition',model='facebook/wav2vec2-base-960h')
```

```python
from datasets import load_dataset,Audio
```

```python
dataset = load_dataset('PolyAI/minds14',name='en-US',split='train')
```

```python
index = 0
for col in dataset.column_names:
  if col == 'audio':
    print(col,dataset[index][col]['array'])
    continue
  print(col,dataset[index][col])
  print('-'*5)
```

```python
print(dataset[index]['audio']['array'].shape)
```

```python
speech_recognizer.feature_extractor
```

```python
import numpy as np
from scipy.io.wavfile import write
from IPython.display import Audio

audio_data = dataset[index]['audio']['array']
sample_rate = speech_recognizer.feature_extractor.sampling_rate

sample_rate = 960

# Normalize the audio data to 16-bit PCM format if needed
audio_data_normalized = np.int16(audio_data / np.max(np.abs(audio_data)) * 32767)

write('output_audio.wav', sample_rate, audio_data_normalized)

# Play the audio directly in Jupyter (or similar environments)
Audio(data=audio_data_normalized, rate=sample_rate)
```
